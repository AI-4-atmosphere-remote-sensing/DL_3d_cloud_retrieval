# 10m Nakajima and King

```python
#r_data = np.empty((4000,2,12,6,4096), dtype=float) # entire reflectance dataset
cot_data=np.empty((4000,4096),dtype=float) #entire cot dataset
cot=np.empty((4000,4096),dtype=float) #ground truth
re_data=np.empty((4000,4096),dtype=float) #entire effective radius 
re=np.empty((4000,4096),dtype=float) #ground truth
for i in range(4000):#4000 profiles
    fname = "D:\\Code\\Jupyter\\climate_project\\New Data V3\\retrieved_cot_re_VZA_0_SZA_60/profile_%05d.hdf5"%(i+1)
    hf = h5py.File(fname, 'r')
  #  r_data[i,:,:,:,:] = np.array(hf.get("reflectance"))
    cot_data[i,:]=np.array(hf.get('rertieved_COT(VZA=0,SZA=60)'))
    cot[i,:]=np.array(hf.get('tau'))
    re_data[i,:]=np.array(hf.get('rertieved_CER(VZA=0,SZA=60)'))
    re[i,:]=np.array(hf.get('re'))
    hf.close()
```
